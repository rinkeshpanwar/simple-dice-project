```python
import random
```

```python
class dice:
    average_previous = 0
    score_list = []
    #average = 0
    avg = 0
    number_of_side = 0
    number_of_dice = 0
    rolled_sum = 0
    bonus = 0
    per = 0
    maxscore = 0
    def __init__(self,name,rollno):
        self.name = name
        self.rollno = rollno
        print('1001-W2020 Final Project By Sample Solution, Student Number'+str(rollno))
        print('Welcome to my dice game, good luck!')
    def faces(self):
        i = 0
        while(i == 0 ):
            try:
                face = int(input('Enter # of faces [2-20]'))
                if (face < 2 or face > 20):
                    print('''
                    I'm sorry that isn't a valid positive integer, please try again.
                    ''')
                else:
                    self.face = face
                    self.number_of_side = face
                    i=i+1
            except:
                print('''
                I'm sorry that isn't a valid positive integer, please try again.
                ''')
    def dices(self):
        i = 0
        while( i == 0):
            try:
                dice = int(input('Enter # of dices [3-6]'))
                if (dice < 3 or dice > 6 ):
                    print('''
                    Sorry you can only use atleast 3 and atost 6 dice
                    ''')
                else:
                    self.dice = dice
                    self.number_of_dice = dice
                    i = i+1
            except:
                print('''
                sorry that's not a valid number try again....
                ''')
    def rollDie(self):
        for i in range(self.dice):
            temp = random.randrange(start=1,stop=self.face)
            self.score_list.append(temp)
    def calculate(self):
        temp = 0
        j = 0
        for i in self.score_list:
            temp = temp + i
            j = j + 1
        avg = round(temp/j)
        self.average_previous = self.avg
        self.avg = avg
        self.rolled_sum = temp
        max_score = self.dice*self.face
        self.maxscore = max_score
        self.per = temp/max_score
        print('This die sum to '+str(temp)+' and has an average value of '+str(avg))
        print('Percentage = '+str(temp/max_score))
    def pattern1(self):
        bonus = 10
        value = self.score_list[0]
        counter = 0
        if (self.face >= 4):
            for i in self.score_list:
                if i != value:
                    counter = 1
            if(counter == 0):
                print('Pattern 1 matched in your roll ' + str(self.score_list))
                self.bonus = bonus + self.bonus
            else:
                print('Pattern 1 didnt matched in your roll'+str(self.score_list))
        else:
            print('Pattern 1 didnt matched in your roll'+str(self.score_list))
    def pattern2(self):
        bonus = 15
        counter = 0
        if (self.maxscore>=20):
            total = 0
            for i in self.score_list:
                total = total + i
            for i in range(2,total):
                if total%i == 0:
                    counter = 1
            if (counter == 0):
                print('Pattern 2 matched in your roll ' + str(self.score_list))
                self.bonus = bonus + self.bonus
            else:
                print('Pattern 2 didnt matched in your roll'+str(self.score_list))
        else:
            print('Pattern 2 didnt matched in your roll'+str(self.score_list))
    def pattern3(self):
        bonus = 5
        half_dice = self.dice/2
        rounded_avg = self.avg
        if(self.dice >=5):
            if(half_dice >= self.avg):
                print('Pattern 3 matched in your roll ' + str(self.score_list))
                self.bonus = bonus + self.bonus
            else:
                print('Pattern 3 didnt matched in your roll'+str(self.score_list))
        else:
            print('Pattern 3 didnt matched in your roll'+str(self.score_list))
    def pattern4(self):
        bonus = 8 
        counter = 0
        total_element = 0
        if(self.dice >4 and self.face > self.dice):
            for i in self.score_list:
                total_element = total_element+1
            for i in range(total_element):
                for j in range(total_element):
                    if(i==j):
                        pass
                    else:
                        if(self.score_list[i]==self.score_list[j]):
                            counter = 1
            if(counter==0):
                print('Pattern 4 matched in your roll ' + str(self.score_list))
                self.bonus = bonus + self.bonus
            else:
                print('Pattern 4 didnt matched in your roll'+str(self.score_list))
        else:
            print('Pattern 4 didnt matched in your roll'+str(self.score_list))
    def pattern5(self):
        bonus=1
        if self.bonus == 0:
            print('Pattern 5 matched in your roll ' + str(self.score_list))
            self.bonus = bonus + self.bonus
        else:
            print('Since you already matched a pattern, pattern5 did not match')
    def reroll(self):
        print('You have '+str(self.dice))
        i = 0
        while(i < self.dice):
            try:
                print('Reroll die'+str(i+1)+'(was'+str(self.score_list[i])+') yes , no')
                inp = input()
                if(inp.lower()=='yes'):
                    temp = random.randrange(start=1,stop=self.face)
                    self.score_list[i]=temp
                    i=i+1
                elif(inp.lower()=='no'):
                    i = i+1
                else:
                    print(''' I'm sorry, the choices are ['yes','no'].please pick one of them ''')
            except:
                print(''' I'm sorry, the choices are ['yes','no'].please pick one of them ''')
```
